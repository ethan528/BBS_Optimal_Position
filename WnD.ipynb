```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')

jan = pd.read_csv('./CALL_NDELIVERY_01MONTH.csv')
feb = pd.read_csv('./CALL_NDELIVERY_02MONTH.csv')
mar = pd.read_csv('./CALL_NDELIVERY_03MONTH.csv')
apr = pd.read_csv('./CALL_NDELIVERY_04MONTH.csv')
may = pd.read_csv('./CALL_NDELIVERY_05MONTH.csv')
jun = pd.read_csv('./CALL_NDELIVERY_06MONTH.csv')
jul = pd.read_csv('./CALL_NDELIVERY_07MONTH.csv')
oug = pd.read_csv('./CALL_NDELIVERY_08MONTH.csv')
sep = pd.read_csv('./CALL_NDELIVERY_09MONTH.csv')
oct = pd.read_csv('./CALL_NDELIVERY_10MONTH.csv')
nov = pd.read_csv('./CALL_NDELIVERY_11MONTH.csv')
dec = pd.read_csv('./CALL_NDELIVERY_12MONTH.csv')
rn = pd.read_csv('./rn_강수량.csv', encoding='ansi')
fsum = pd.read_csv('./STCS_체감온도_여름.csv', encoding='ansi')
fwin = pd.read_csv('./STCS_체감온도_겨울.csv', encoding='ansi')
ta = pd.read_csv('./ta_온도.csv', encoding='ansi')
```

```python
def dash_del(df, col):
    o = df[col]

    for i in range(len(o)):
        o[i] = int(o[i].replace('-', ''))
    
    df[col] = list(o)
    df = df.astype({col:'int'})
    
    return df

rn = rn.rename(columns={'날짜':'일자'})
ta = ta.rename(columns={'날짜':'일자'})
weather = [rn, fsum, fwin, ta]

for weather in weather:
    weather = dash_del(weather, '일자')

months = [nov, dec, jan, feb, mar, apr, may, jun, jul, oug, sep]
wnd = oct

for month in months:
    wnd = wnd.append(month)
```
